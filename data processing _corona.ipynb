```python
import pandas as pd
import glob
import os

input_file = r'D:/Programming/Github/Bigdata/COVID-19/csse_covid_19_data/csse_covid_19_daily_reports' # csv파일들이 있는 디렉토리 위치
geo_file = r'./geo.csv'
output_file = r'./result.csv' # 병합하고 저장하려는 파일명
output_file_axis = r'./result_axis.csv'

allFile_list = glob.glob(os.path.join(input_file, '*.csv')) # glob함수로 sales_로 시작하는 파일들을 모은다



allData = [] # 읽어 들인 csv파일 내용을 저장할 빈 리스트를 하나 만든다


for file in allFile_list:
    df = pd.read_csv(file) # for구문으로 csv파일들을 읽어 들인다
    df['State']=df['Province/State'].fillna('')
    df2 = pd. read_csv(geo_file)
#     df['region']= df['Province/State'] + ', ' +df['Country/Region']
    df['region']= df[['State', 'Country/Region']].apply(lambda x: ''.join(x), axis=1)
    
#     df= df.drop(['Province/State','Country/Region'],1)
    df['filedate'] = os.path.basename(file) 
    df['Date']  = df['filedate'].str.replace(pat=r'.csv', repl= r' ', regex=True)
    df=pd.merge(df, df2, left_on='region',right_on='region2', how='left')
    
    
    allData.append(df) # 빈 리스트에 읽어 들인 내용을 추가한다
    

dataCombine = pd.concat(allData, axis=0, ignore_index=True) # concat함수를 이용해서 리스트의 내용을 병합
# axis=0은 수직으로 병합함. axis=1은 수평. ignore_index=True는 인데스 값이 기존 순서를 무시하고 순서대로 정렬되도록 한다.

dataCombine.drop(['filedate','Last Update'], axis='columns', inplace=True)
# dataCombine.drop(['filedate'], axis='columns', inplace=True)
# dataCombine.drop(['region2','Latitude','Longitude'], axis='columns', inplace=True)
dataCombine.to_csv(output_file, index=False) # to_csv함수로 저장한다. 인데스를 빼려면 False로 설정
```
